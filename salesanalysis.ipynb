```python
import pandas as pd
```

**merge all 12 months data into a single file**

```python
data=pd.concat(map(pd.read_csv, ['Sales_April_2019.csv','Sales_August_2019.csv','Sales_December_2019.csv','Sales_February_2019.csv',
        'Sales_January_2019.csv','Sales_July_2019.csv','Sales_June_2019.csv','Sales_March_2019.csv','Sales_May_2019.csv'
        ,'Sales_November_2019.csv','Sales_October_2019.csv','Sales_September_2019.csv']))
data
```

```python
alldata=pd.DataFrame(data)
alldata.head()
```

**Clean The data**

```python
nan=alldata[alldata.isna().any(axis=1)]
alldata=alldata.dropna(how='all')
alldata
```

**Clean all  'or'**

```python
alldata=alldata[alldata['Order Date'].str[0:2]!='Or']
```

**add columns**

**1.add month column**


```python
alldata['Month']=alldata['Order Date'].str[0:2]
alldata['Month']=alldata['Month'].astype('int32')
alldata.head()
```

```python
#add a column with sales = quantity ordered * price each
alldata['Quantity Ordered']=pd.to_numeric(alldata['Quantity Ordered'])
alldata['Price Each']=pd.to_numeric(alldata['Price Each'])
alldata['Sales']=alldata['Quantity Ordered']*alldata['Price Each']
alldata
```

Best month For sales and how much was earned the month

```python
alldata.groupby('Month').Sales.sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
```

```python
months=range(1,13)
plt.bar(months,alldata.groupby('Month').Sales.sum())
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

What city had the highest number of sales

```python
#add a city column
alldata['City']=alldata['Purchase Address'].apply(lambda x:x.split(',')[1]+' '+x.split(',')[2].split(' ')[1])
alldata.head()
```

```python
city=alldata.groupby('City'). Sales.sum()
city
```

```python
City=[city for city ,df in alldata.groupby('City')]
plt.bar(City,alldata.groupby('City').Sales.sum())
plt.xticks(City,rotation='vertical')
plt.ylabel('Sales in USD ($)')
plt.xlabel('City')
plt.show()
```

Best time to display ads to maximize the customer's buying likelihood?

```python
alldata['Order Date']=pd.to_datetime(alldata['Order Date'])
alldata['Hour']=alldata['Order Date'].dt.hour
alldata['Minute']=alldata['Order Date'].dt.minute
alldata
```

```python
alldata.groupby('Hour').Sales.sum()
```

```python
hours = [hour for hour ,df in alldata.groupby('Hour')]
plt.plot(hours,alldata.groupby('Hour').Sales.sum())
plt.xticks(hours)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Hour of the day')
plt.show()
#Around 10am to 12noon and 6pm to 8pm
```

What products are sold together the most?

```python
df=alldata[alldata['Order ID'].duplicated(keep=False)]
df['Grouped']=df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))
df=df[['Order ID','Grouped']].drop_duplicates()
df
```

```python
from itertools import combinations
from collections import Counter
count=Counter()
for row in df['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
for key,value in count.most_common(10):
    print(key,value)
```
